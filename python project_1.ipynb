```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"project Data new.csv", encoding= 'unicode_escape')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(["Status", "unnamed1"], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
plt.Figure(figsize=(10,6))

Color = ['#e74c3c', '#ee796d']


A1 = sns.countplot(x="Gender",data=df , palette=Color,edgecolor="black",linewidth=1.7)

plt.title("Gender Wise Count",fontsize=22,family="Georgia",fontweight="bold")

plt.xlabel("Gender",fontsize=16,family="Georgia",fontweight="bold")

plt.ylabel("Gender Count M / L" ,fontsize = 16, family="Arial",fontweight="bold")

for x in A1.containers:
    A1.bar_label(x)
plt.show()
```

```python
plt.Figure(figsize=(10,6))
Color = ['#e74c3c', '#ee796d', '#f2a29b', '#f6cfc6', '#fbeae8']

A2 = sns.countplot(x="Age Group", data=df, hue="Gender", palette=Color, edgecolor="black",linewidth=1.7)
plt.title("Age_Group Wise Gender Distribution",fontsize=22,family="Georgia",fontweight="bold")

plt.xlabel("Age Group",fontsize=16,family="Georgia",fontweight="bold")

plt.ylabel("Count of Age Group" ,fontsize = 16, family="Arial",fontweight="bold")
for i in A2.containers:
    A2.bar_label(i)
```

```python
df.rename(columns={"Amount":"Profit"}, inplace=True)
```

```python
df
```

```python
Color = ['#e74c3c', '#ee796d', '#f2a29b', '#f6cfc6', '#fbeae8']
A3 = sns.barplot(x="Zone", y="Profit", data=df, ci=None, palette=Color, edgecolor="black",linewidth=1.7)
plt.title("Zone Wise Profit",fontsize=22,family="Georgia",fontweight="bold")

plt.xlabel("Zone",fontsize=16,family="Georgia",fontweight="bold")

plt.ylabel("Profit" ,fontsize = 16, family="Arial",fontweight="bold")
for i in A3.containers:
    A3.bar_label(i)
```

```python
zone_count = df["Zone"].value_counts()
zone_count
```

```python
Color = ['#e74c3c', '#ee796d', '#f2a29b', '#f6cfc6', '#fbeae8']
plt.pie(zone_count, labels=zone_count.index, autopct="%1.f%%", startangle=310, colors=Color)
plt.title("Zone Wise Count",fontsize=22,family="Georgia",fontweight="bold")
```

```python
Gender_count = df["Gender"].value_counts()
Gender_count
```

```python
Color = ['#e74c3c', '#ee796d']
plt.pie(Gender_count, labels=Gender_count.index, autopct="%1.f%%", colors=Color)
plt.title("Count of Gender",fontsize=22,family="Georgia",fontweight="bold")
```

```python
state_order = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by = "Orders", ascending=False)
state_order
state_order1 = state_order.head(5) 
```

```python
Color = ['#e74c3c', '#ee796d', '#f2a29b', '#f6cfc6', '#fbeae8']
A4 = sns.barplot(x="State", y="Orders", data=state_order1, ci=None, palette=Color, edgecolor="black",linewidth=1.7)
for i in A4.containers:
    A4.bar_label(i)
plt.xticks(rotation = 90)
plt.title("State Wise Order",fontsize=22,family="Georgia",fontweight="bold")

plt.xlabel("State",fontsize=16,family="Georgia",fontweight="bold")

plt.ylabel("Orders" ,fontsize = 16, family="Arial",fontweight="bold")
```

```python
Count_cat = df["Product_Category"].value_counts().reset_index()
Count_cat
Count_cat1 = Count_cat.head(5)
```

```python
Color = ['#e74c3c', '#ee796d', '#f2a29b', '#f6cfc6', '#fbeae8']
A5 = sns.barplot(x="Product_Category", y="count", data=Count_cat1, palette=Color, edgecolor="black",linewidth=1.7)
for i in A5.containers:
    A5.bar_label(i)
plt.xticks(rotation = 90)
plt.title("Product_Categoery Wise Count",fontsize=22,family="Georgia",fontweight="bold")
```
